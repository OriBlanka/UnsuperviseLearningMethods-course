### 7.1. GMM Algorithm

The GMM algorithm aims to maximize the (log) likelihood objective:
$$\arg\max_{\left\{ \left(w_{k},\boldsymbol{\mu}_{k},\boldsymbol{\Sigma}_{k}\right)\right\} _{k=1}^{K}}f=\arg\max_{\left\{ \left(w_{k},\boldsymbol{\mu}_{k},\boldsymbol{\Sigma}_{k}\right)\right\} _{k=1}^{K}}\sum_{i=1}^{N}\log\left(\sum_{k=1}^{K}w_{k}\mathcal{N}\left(\boldsymbol{x}_{i};\boldsymbol{\mu}_{k},\boldsymbol{\Sigma}_{k}\right)\right)$$

In this section you should implement:

1. `InitGmm()` - Given a data set and number of clusters it sets the initialization of the `GMM` algorithm:  
   - `mμ` - Should be initialized by the [`K-Means++`](https://en.wikipedia.org/wiki/K-means%2B%2B) algorithm.
   - `tΣ` - Should be initialized by diagonal matrices with the data variance on the diagonal (A scalar matrix).
   - `vW` - Should be initialized with a uniform distribution.  
2. `CalcGmmObj()` - Given a data set and set of parameters it calculate the value of the GMM objective function.
3. `GMM()` - Given a data set and initialization applies the GMM algorithm.  
The stopping criteria should be number of iterations or minimal improvement in the objective function.

* <font color='brown'>(**#**)</font> Implementation should be efficient (Memory and operations). Total run time expected to be **less than 60 seconds**.
* <font color='brown'>(**#**)</font> You may use the `scipy.stats.multivariate_normal` class.



```python
def InitGmm(mX: np.ndarray, K: int, seedNum: int = 123) -> np.ndarray:
    '''
    GMM algorithm initialization.
    Args:
        mX          - Input data with shape N x d.
        K           - Number of clusters.
        seedNum     - Seed number used.
    Output:
        mμ          - The initial mean vectors with shape K x d.
        tΣ          - The initial covariance matrices with shape (d x d x K).
        vW          - The initial weights of the GMM with shape K.
    Remarks:
        - Given the same parameters, including the `seedNum` the algorithm must be reproducible.
        - mμ Should be initialized by the K-Means++ algorithm.
    '''

    mμ = InitKMeans(mX, K, initMethod= 1, seedNum= seedNum)
    vW = np.ones(K) / K
    tΣ = np.zeros((mX.shape[1], mX.shape[1], K))
    for i in range(K):
        tΣ[:, :, i] = np.eye(mX.shape[1])
    return mμ, tΣ, vW
    
#===============================================================#
```

```python
#===========================Fill This===========================#
def CalcGmmObj(mX: np.ndarray, mμ: np.ndarray, tΣ: np.ndarray, vW: np.ndarray) -> float:
    '''
    GMM algorithm objective function.
    Args:
        mX          - The data with shape N x d.
        mμ          - The initial mean vectors with shape K x d.
        tΣ          - The initial covariance matrices with shape (d x d x K).
        vW          - The initial weights of the GMM with shape K.
    Output:
        objVal      - The value of the objective function of the GMM.
    Remarks:
        - A
    '''

    objVal = 0
    for i in range(mX.shape[0]):
        for j in range(mμ.shape[0]):
            objVal += -np.log(vW[j] * MVN.pdf(mX[i], mμ[j], tΣ[:, :, j]))
    return objVal
#===============================================================#
```

```python
def GMM(mX: np.ndarray, mμ: np.ndarray, tΣ: np.ndarray, vW: np.ndarray, numIter: int = 1000, stopThr: float = 1e-5) -> np.ndarray:
    '''
    GMM algorithm.
    Args:
        mX          - Input data with shape N x d.
        mμ          - The initial mean vectors with shape K x d.
        tΣ          - The initial covariance matrices with shape (d x d x K).
        vW          - The initial weights of the GMM with shape K.
        numIter     - Number of iterations.
        stopThr     - Stopping threshold.
    Output:
        mμ          - The final mean vectors with shape K x d.
        tΣ          - The final covariance matrices with shape (d x d x K).
        vW          - The final weights of the GMM with shape K.
        vL          - The labels (0, 1, .., K - 1) per sample with shape (N, )
        lO          - The objective function value per iterations (List).
    Remarks:
        - The maximum number of iterations must be `numIter`.
        - If the objective value of the algorithm doesn't improve by at least `stopThr` the iterations should stop.
    '''

    lO = []
    vL = np.zeros(mX.shape[0])
    for i in range(numIter):
        # Step 1
        pXI_K = np.zeros((mX.shape[0], mμ.shape[0]))
        for j in range(mX.shape[0]):
            pXI_K[j, :] = [vW[k] * MVN.pdf(mX[j], mμ[k], tΣ[:, :, k] ) for k in range(mμ.shape[0])]
            pXI_K[j, :] /= pXI_K[j, :].sum()

        vL = pXI_K.argmax(axis=1)
        # Step 2
        for k in range(mμ.shape[0]):
            nK = pXI_K[:, k].sum()
            vW[k] = nK / mX.shape[0]
            mμ[k] = 1/nK * (pXI_K[:, k].reshape(-1, 1) * mX).sum(axis=0)
            tΣ[:, :, k] = 1/nK * (pXI_K[:, k].reshape(-1, 1, 1) * (mX - mμ[k]).reshape(-1, 2, 1) * (mX - mμ[k]).reshape(-1, 1, 2)).sum(axis=0)
            
        lO.append(CalcGmmObj(mX, mμ, tΣ, vW))
        if i > 0 and abs(lO[-1] - lO[-2]) < stopThr:
            break
    return mμ, tΣ, vW, vL, lO
```

### 7.3. The Covariance Matrix of the GMM Model

The `GaussianMixture` class of SciKit Learn has the option `covariance_type` with the options: `full`, `tied`, `diag`, `spherical`.  
In class the model assumed full degree of freedom, namely `full`.  
Explain in what cases should one use `tied` / `diag` / `spherical`.  
Address both properties of the data and the parameters of the model.

### 7.3. Solution
1. full (Default):

    When to Use: Use the full option when you have no prior knowledge or assumption about the structure of the data's covariance. This is the most flexible option, allowing each component of the GMM to have its own arbitrary covariance matrix.

    Data Property: Well-suited for data where the clusters have different shapes and orientations, and the covariance structure varies significantly among clusters.

    Model Parameterization: The full covariance type has the most parameters to estimate, making it the most expressive but also the most computationally expensive. It can capture complex relationships within the data.

2. tied:

    When to Use: Use the tied option when you assume that all components share the same covariance matrix. This reduces the complexity of the model and can help with overfitting when you have limited data.

    Data Property: Suitable when clusters have similar shapes and orientations, and you want to constrain the model to assume that they all share a common covariance structure.

    Model Parameterization: tied covariance type estimates a single covariance matrix for all components, significantly reducing the number of parameters compared to full. This can be beneficial when you have limited data.

3. diag:

    When to Use: Use the diag option when you assume that the covariance matrices for each component are diagonal, meaning there is no correlation between different features. This is less restrictive than spherical but still simplifies the model.

    Data Property: Suitable when you believe that features are mostly independent within clusters, and you want to capture different variances along each dimension.

    Model Parameterization: diag covariance type estimates a diagonal covariance matrix for each component, which is less complex than full but more expressive than spherical. It strikes a balance between model complexity and flexibility.

4. spherical:

    When to Use: Use the spherical option when you assume that all clusters are spherical and have the same variance along all dimensions. This is the most restrictive but simplest option.

    Data Property: Suitable when you have strong evidence that clusters are spherical and equally spread in all directions from their centroids.

    Model Parameterization: spherical covariance type estimates a single scalar variance for each component, making it the simplest and least parameterized option. It is suitable for scenarios where simplicity and interpretability are essential, but it may not capture complex covariance structures.

### 7.4. Solution
1. Parameters for GMM to Mimic K-Means:

    To make a Gaussian Mixture Model (GMM) behave like K-Means, you need to set its parameters as follows:

    Covariance Matrices: Set the covariance matrices to be diagonal matrices with equal diagonal elements (i.e., spherical covariances). In this case, each Gaussian component will have the same variance in all dimensions, effectively making them isotropic.

    Weights: Set the weights of all Gaussian components to be equal, meaning that each component contributes equally to the mixture. This ensures that all clusters have the same importance in the GMM as in K-Means.

    Number of Components: The number of Gaussian components in the GMM should be equal to the number of clusters (K) in K-Means.

    By following these parameter settings, the GMM will essentially behave like the K-Means algorithm.
    

2. Data Transformation for K-Means-like Behavior in GMM:

    The primary difference between K-Means and GMM is how they model cluster shapes. K-Means assumes spherical clusters, while GMM allows for elliptical and more flexible cluster shapes due to its covariance matrices. To make GMM perform like K-Means on a given dataset:

    Rotate the Data:

    You can rotate the data using a rotation matrix to make it spherical or isotropic. The steps to do this are as follows:

    Compute the covariance matrix of the original data.
    Calculate its eigenvalues and eigenvectors.
    Form a rotation matrix using the eigenvectors.
    Multiply the original data by the rotation matrix.
    The rotated data will have its principal axes aligned with the coordinate axes, making the GMM behave like K-Means on this transformed data.
